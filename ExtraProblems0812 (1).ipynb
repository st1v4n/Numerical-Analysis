```python
### Zadacha 1

height_nodes = np.array([0.0, 1.525, 3.05, 4.575, 6.1])
values = np.array([1.0, 0.8617, 0.7385, 0.6292, 0.5328])
```

```python
def lagr(x, target_x, all_nodes):
    up_side = 1.0
    down_side = 1.0
    for i in all_nodes:
        if i == target_x:
            continue
        up_side *= (x - i)
        down_side *= (target_x - i)
    return up_side / down_side

def lagr_poly(x, nodes, values):
    result = 0.0
    for i in range(len(values)):
        result += values[i] * lagr(x, nodes[i], nodes)
    return result

print("Value in height = 3km : " + str(lagr_poly(3.0, height_nodes,values))) # passed
plt.scatter(height_nodes, values, color = 'red')
x_axis = np.linspace(0.0, 6.1, 200)
plt.plot(x_axis, lagr_poly(x_axis, height_nodes, values))
plt.show()
```

```python
# Zadacha 2
nodes = np.array([0.0, 0.0, 1.0, 1.0, 2.0])
values = np.array([1.0, 0, 2, 6, 21])
dict_nodes = {0.0 : 1.0, 1.0 : 2, 2.0 : 21}
def devided_difference(nodes, values):
    if len(nodes) == 1:
        return dict_nodes[nodes[0]]
    if nodes[0] == nodes[len(nodes) - 1]:
        factoriel = 1.0
        for j in range(2, len(nodes)):
            factoriel *= j
        return values[len(nodes) - 1] / factoriel
    dev = (devided_difference(nodes[1:], values[1:]) - devided_difference(nodes[0:-1], values[0:-1])) / (nodes[-1] - nodes[0])
    return dev

def newton_poly(x, nodes, values):
    addition = 1.0
    result = 0.0
    for i in range(len(nodes)):
        dev = devided_difference(nodes[0:i+1], values[0:i+1]) * addition
        result += dev
        addition *= (x-nodes[i])
    return result
print(newton_poly(1.5, nodes, values)) # passed
```

```python
nodes = np.array([1.0,2,3,4,5])
values = np.array([3.0, 128, 115, 1050, 13500])

def exp_row(t0):
    result = [1]
    for i in range(1, nodes.size):
        result.append(math.e**(i*t0))
    return result
    
def exp_matrix(nodes):
    result = []
    for i in nodes:
        result.append(exp_row(i))
    return np.array(result)
    
def exp_poly(nodes, x):
    result = nodes[0]
    for i in range(1, nodes.size):
        result += nodes[i] * (math.e**(i*x))
    return result

def alg_row(t0):
    result = [1]
    for i in range(1, nodes.size):
        result.append(t0**i)
    return result

def alg_matrix(nodes):
    result = []
    for i in nodes:
        result.append(alg_row(i))
    return np.array(result)

def alg_poly(nodes, x):
    result = nodes[0]
    for i in range(1, nodes.size):
        result += nodes[i] * (x**i)
    return result

matrix = alg_matrix(nodes)
solution_of_system = np.linalg.solve(matrix, values)
x_axis = np.linspace(1,5,200)
plt.scatter(nodes,values,color='red')
plt.plot(x_axis, alg_poly(solution_of_system, x_axis))
plt.show()
print(alg_poly(solution_of_system, 4.9))
```
