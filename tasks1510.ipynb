```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)
```

```python
x_sin_axis = np.linspace(0,2*math.pi,1000)
```

```python
plt.plot(x_sin_axis,f(x_sin_axis))
plt.legend(["sin(x)"])
plt.show()
```

```python
def lagr(x,x0,x1,x2,x3):
    return 0.5*(((x-x0)*(x-x2)*(x-x3))/((x1-x0)*(x1-x2)*(x1-x3))) + (math.sqrt(3)/2)*(((x-x0)*(x-x1)*(x-x3))/((x2-x0)*(x2-x1)*(x2-x3))) + 1*(((x-x0)*(x-x1)*(x-x2))/((x3-x0)*(x3-x1)*(x3-x2)))
```

```python
math.sin(math.pi / 5)
```

```python
plt.plot(x_sin_axis,lagr(x_sin_axis,0,math.pi/6,math.pi/3,math.pi/2))
plt.legend("lagr(x)")
plt.show()
```

```python
plt.plot(x_sin_axis,f(x_sin_axis), x_sin_axis, lagr(x_sin_axis,0,math.pi/6,math.pi/3,math.pi/2))
plt.legend(["sin(x)", "lagr(x)"])
plt.show()
```

```python
plt.plot(x_sin_axis, abs(f(x_sin_axis) - lagr(x_sin_axis,0,math.pi/6,math.pi/3,math.pi/2)))
plt.legend(['abs greshka'])
plt.show()
```
