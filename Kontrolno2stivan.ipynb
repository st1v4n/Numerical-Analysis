```python
import math
import numpy as np
import matplotlib.pyplot as plt
from sympy import solve, Eq, diff, symbols, integrate
```

```python
# Задача 1

nodes = np.array([3.0, 4, 5, 7, 8, 10, 11, 12])
values = np.array([1.6, 3.6, 4.4, 3.4, 2.5, 2.8, 3.8, 4.6])

a, b, c, d = symbols('a, b, c, d')

def f(x):
    return a + b*x + c*(x**2) + d*(x**3)

phi = 0

for i in range(nodes.size):
    phi += (f(nodes[i]) - values[i])**2

diffA = diff(phi, a)
diffB = diff(phi, b)
diffC = diff(phi, c)
diffD = diff(phi, d)

solution = solve([Eq(diffA, 0), Eq(diffB, 0), Eq(diffC, 0), Eq(diffD, 0)])

def f_solution(x):
    return solution[a] + solution[b]*x + solution[c]*(x**2) + solution[d]*(x**3)

plt.scatter(nodes, values, color='red')
x_axis = np.linspace(3, 12, 200)
plt.plot(x_axis, f_solution(x_axis))
plt.legend(['Actual points', 'Generated function'])
plt.show()
```

```python
# => трябва ни поне n = 366, За да успеем!!!
n = 366
a = 1
b = 3
h = (b-a)/n

def f(x):
    return np.sin(x)**2

nodes = []

for i in range(0,n+1):
    nodes.append(a + i*h)

sum = 0

for i in range(1, n+1):
    sum += (f(nodes[i-1]) + f(nodes[i]))

result = ((b-a)/(2*n))*sum
print("Result: " + str(result))
```
